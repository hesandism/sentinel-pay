```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../data/fraudTrain.csv")
test  = pd.read_csv("../data/fraudTest.csv")

# ── Row counts ───────────────────────────────────────────────
print("=" * 45)
print(f"  Train rows : {len(train):,}")
print(f"  Test rows  : {len(test):,}")
print(f"  Total rows : {len(train) + len(test):,}")
print("=" * 45)

# ── Fraud rate ───────────────────────────────────────────────
for name, df in [("Train", train), ("Test", test)]:
    total     = len(df)
    frauds    = df["is_fraud"].sum()
    fraud_pct = frauds / total * 100
    print(f"\n{name}:")
    print(f"  Legit transactions : {total - frauds:,}")
    print(f"  Fraud transactions : {frauds:,}")
    print(f"  Fraud rate         : {fraud_pct:.2f}%")

print("\n✅ Fraud rate should be ~0.5% — confirming class imbalance.")
```
